```python
import cv2
import os
import sys
import shutil
from PyQt5 import QtGui, QtWidgets
```

```python
# open the video file
def openfile_dialog():
    app = QtWidgets.QApplication([dir])
    fname = QtWidgets.QFileDialog.getOpenFileName(None, "Select a file...", '.', filter="All files (*)")
    return str(fname)
```

```python
# manage directory
def createDirectory(path):
    if not os.path.exists(path):
        os.makedirs(path)
        
def removeDirectory(path):
    if os.path.exists(path):
        shutil.rmtree(path)
        print("Files have been removed!")
```

```python
#Initialization

filename = openfile_dialog()
filename = filename.split('\'')[1].split('\'')[0]


FilePath = filename.split('.')[0]
print(FilePath)
if os.path.exists(FilePath):
    print("Video has been processed!")
    removeDirectory(FilePath)

createDirectory(FilePath)
print("Directory created!")
```

```python
def writeFrames(videoPath, frameCount, image):
      cv2.imwrite(videoPath + "frame%d.jpg" % frameCount, image) 
```

```python
#Create corresponding video folder
def writeVideo(videoPath, videoCount, img_array, fps):
    
    #Create videowriter
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    height, width, layers = img_array[0].shape
    size = (width,height)
    videoOut = cv2.VideoWriter(videoPath + "video%d.avi" % videoCount, fourcc, fps, size) 
    #Write frames to the video
    if videoOut.isOpened():
        for i in range(len(img_array)):
            videoOut.write(img_array[i])
    else:
        print("Videowriter created failed!")
    
    videoOut.release()
    VIDEOFLAG = False
    return VIDEOFLAG
```

```python
#Read next frame
def nextFrame(vidcap, frameCount):
    #Close the image window
    cv2.destroyAllWindows()
    #Forward to next frame
    success,image = vidcap.read()
    if success:
        frameCount += 1
        return frameCount, image
    else:
        print("Vieo processing finished!")
```
