```python
import cv2
import os
import sys
from PyQt5 import QtGui, QtWidgets
```

```python
def openfile_dialog():
    app = QtWidgets.QApplication([dir])
    fname = QtWidgets.QFileDialog.getOpenFileName(None, "Select a file...", '.', filter="All files (*)")
    return str(fname)
```

```python
def createDirectory(path):
    if not os.path.exists(path):
        os.makedirs(path)
        
def removeDirectory(path):
    if os.path.exists(path):
        os.rmdir(path)
        print("Files have been removed!")
```

```python
filename = openfile_dialog()
filename = filename.split('\'')[1].split('\'')[0]
print(filename)
```

```python
LabelledFilePath = filename.split('.')[0]
print(LabelledFilePath)
if os.path.exists(LabelledFilePath):
    print("Video has been processed!")
    removeDirectory(LabelledFilePath)

createDirectory(LabelledFilePath)
print("Directory created!")
```

```python
def writeFrames(videoPath, frameCount, image):
      cv2.imwrite(videoPath + "frame%d.jpg" % frameCount, image) 
```

```python
def processVideo(filename, LabelledFilePath):
    vidcap = cv2.VideoCapture(filename)
    videoCount = 0
    frameCount = 0
    fps = 30
    
    print(filename)
    print(LabelledFilePath)
    print("Start video processing...")
    print("Press g key if the frame is good.")
    print("Press space key to cotinue cropping.")
    print("Press b key to stop cropping if the frame is bad.")
    print("Press c key to end the processing.")
    input("Press any key to continue...")
    
    success,image = vidcap.read()
    
    if success:
        print("\nVideo loading succeeded!")  
        VIDEOFLAG = False
    else:
        print("Video loading failed!")
        
    while success:
        cv2.imshow("Video%d" % videoCount, image)
        k = cv2.waitKey(0)
             
        if k == ord('g'):
            videoCount += 1
            frameCount +=1
            #Create corresponding video folder
            createDirectory(LabelledFilePath + '/Video%d' % videoCount)
            videoPath = LabelledFilePath + '/Video%d/' % videoCount
             #Create videowriter
            if not VIDEOFLAG:
                fourcc = cv2.VideoWriter_fourcc(*'XVID')
                videoOut = cv2.VideoWriter(videoPath + "video%d.avi" % videoCount, fourcc, fps, image.shape[:2]) 
                if videoOut.isOpened():
                    VIDEOFLAG = True
                else:
                    print("Videowriter created failed!")
    
            #Write video and frame
            videoOut.write(image)
            writeFrames(videoPath, frameCount, image)
            #Close the image window
            cv2.destroyAllWindows()
            #Forward to next frame
            success,image = vidcap.read()
            if success:
                frameCount += 1
                continue
            else:
                print("Vieo processing finished!")
            
        elif k == ord(' '):
            if VIDEOFLAG:  
                #Write video and frame
                videoOut.write(image)
                print("Recoding...")
                writeFrames(videoPath, frameCount, image)
            else:
                print("Not recoding...")
            #Close the image window
            cv2.destroyAllWindows()
            #Forward to next frame
            success,image = vidcap.read()
            if success:
                frameCount += 1
                continue
            else:
                print("Vieo processing finished!")
        
        elif k == ord('b'):
            #Close the videowriter
            if VIDEOFLAG:
                print("Video ended")
                videoOut.release()
                VIDEOFLAG = False
                #Forward to next frame
                success,image = vidcap.read()
                #Set frameCount to 0
                if success:
                    frameCount = 0
                    continue
                else:
                    print("Vieo processing finished!")
                #Close the image window
                cv2.destroyAllWindows()
            else:
                print("press g key to start recording first or press space key to next frame!")
            
           
            
        
        elif k == ord('c'):
            cv2.destroyAllWindows()
            vidcap.release()
            print("-"*50)
            print("Video processing stopped!")
            break
        
        else:
            print("Wrong Key!")
```

```python
processVideo(filename, LabelledFilePath)
```
